```python
df_exp_merg = pd.read_csv('experiment/merged/exp_merged_pre_and_flight.csv')
df_ctrl_merg = pd.read_csv('control/merged/control_merged_raw.csv')
df_exp_merg.head()
df_exp_merg.tail()
df_ctrl_merg.head()
df_ctrl_merg.tail()

```

```python
cols_to_keep = ['timeStamp',
                'Exp0_OxygenTemp', 'Exp0_OxygenpercentO2', 'Exp0_PhValue',
                'Exp1_OxygenTemp', 'Exp1_OxygenpercentO2',
                'Pressure_LateAccess']
df_ctrl = df_ctrl_merg[cols_to_keep]
df_exp = df_exp_merg[cols_to_keep]
df_ctrl.head(2)
df_ctrl.tail(2)
df_exp.head(2)
df_exp.tail(2)
```

```python
# Transform columns values 
# Control
df_ctrl.loc[:,'Exp0_OxygenTemp'] = df_ctrl.Exp0_OxygenTemp/1000
df_ctrl.loc[:,'Exp1_OxygenTemp'] = df_ctrl.Exp1_OxygenTemp/1000
df_ctrl.loc[:,'Exp0_OxygenpercentO2'] = df_ctrl.Exp0_OxygenpercentO2/1000
df_ctrl.loc[:,'Exp1_OxygenpercentO2'] = df_ctrl.Exp1_OxygenpercentO2/1000
df_ctrl.loc[:,'timeStamp'] = pd.to_datetime(df_ctrl.timeStamp, unit='s')
# Experiment
df_exp.loc[:,'Exp0_OxygenTemp'] = df_exp.Exp0_OxygenTemp/1000
df_exp.loc[:,'Exp1_OxygenTemp'] = df_exp.Exp1_OxygenTemp/1000
df_exp.loc[:,'Exp0_OxygenpercentO2'] = df_exp.Exp0_OxygenpercentO2/1000
df_exp.loc[:,'Exp1_OxygenpercentO2'] = df_exp.Exp1_OxygenpercentO2/1000
df_exp.loc[:,'timeStamp'] = pd.to_datetime(df_exp.timeStamp, unit='s')

```

```python
# Date ranges
# Experiment
print('Start time Experiment:')
min(df_exp.timeStamp)
print('End time Experiment:')
max(df_exp.timeStamp)
# Control
print('Start time Control:')
min(df_ctrl.timeStamp)
print('End time Control:')
max(df_ctrl.timeStamp)
```
