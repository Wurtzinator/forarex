```python
def read_all_csvs_in_folder(path):
    import pandas as pd
    import glob
    import os
    # Create list containing all files names of the current folder
    all_files = glob.glob(path + "/*.csv")
    print('Folder contains this many files: ' + str(len(all_files)))
    # This empty list will be filled with the names of the CSV files we want to append/merge
    li = []
    n_files_appended = 0
    filename_list = []
    
    for filename in all_files:
        print('Reading current file:')
        print(filename)
        # Check whether the CSV file is larger than 0Bytes (if not, it has for sure no data and will break during read in)
        if os.path.getsize(filename) > 0:
            df = pd.read_csv(filename, index_col=None, header=0)
            # Check whether the CSV is empty (this ALSO works when it DOES have headers, but no row data. This is important!)
        #    if not df.empty:
            li.append(df)
            n_files_appended = n_files_appended + 1
            filename_list.append(filename)
    
    # Merge all files in the file list
    frame = pd.concat(li, axis=0, ignore_index=True)
    print('I appended this many files: ' + str(n_files_appended))
    return frame
```

```python
# Read in files and merge them
#PreFlight & Flight
path_pre_and_flight = r'/home/taco/Documents/greta_forarex/data/experiment/raw/exp_raw_pre_and_flight/' # use your path
pre_and_flight = read_all_csvs_in_folder(path_pre_and_flight)

# PostFlight
path_postflight = r'/home/taco/Documents/greta_forarex/data/experiment/raw/exp_raw_postflight/' # use your path
postflight = read_all_csvs_in_folder(path_postflight)

```
